```python
import os
import random
import shutil
import numpy
```

```python
path = 'PokemonData'
all_pokemon= os.listdir(path)
print(all_pokemon)
train_pokemon = "PokemonData/Poke_Train"
test_pokemon = "PokemonData/Poke_Test"
```

```python
def create_test_data(poke_name,train_pokemon,test_pokemon):
    pokemon = os.listdir(train_pokemon+'/'+poke_name)
    os.system('find PokemonData/Poke_Train/ -name ".DS_Store" -type f -delete')
    test_data = random.sample(pokemon,12)
    print(test_data)
    
    for f in test_data:
        shutil.copy(train_pokemon+'/'+poke_name+'/'+f,test_pokemon+'/'+poke_name+'/')
```

```python
for poke in os.listdir(train_pokemon):
    create_test_data(poke,train_pokemon,test_pokemon)
    os.system('find PokemonData/Poke_Train/ -name ".DS_Store" -type f -delete')
```
